```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dfs = [pd.read_csv(f'data/{lg}_2023-2024.csv') for lg in ['Premier-League', 'Liga', 'Bundesliga', 'Serie-A', 'Eredivise', 'Primeira-Liga']]
data = pd.concat(dfs).sort_values(['Year', 'League']).reset_index()
```

```python
win_shares = data.groupby('Year')[['Home_W', 'Draw', 'Away_W']].mean().reset_index()
```

```python
plt.plot(win_shares['Year'], win_shares['Home_W'], 'g-o')
plt.plot(win_shares['Year'], win_shares['Draw'], 'y-o')
plt.plot(win_shares['Year'], win_shares['Away_W'], 'r-o')
```
